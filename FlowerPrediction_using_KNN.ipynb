```python
import pandas as pd
import matplotlib.pyplot as plt
import seaborn as sns
from sklearn.datasets import load_iris
from sklearn.model_selection import train_test_split
from sklearn.neighbors import KNeighborsClassifier
from sklearn.metrics import accuracy_score,confusion_matrix,classification_report
from sklearn.preprocessing import StandardScaler,MinMaxScaler
```

```python
Dataset=load_iris()
```

```python
Dataset
```

```python
Dataset.feature_names
```

```python
x=Dataset.data
y=Dataset.target
```

```python

df=pd.DataFrame(x,columns=Dataset.feature_names)
df
```

```python
y
```

```python
print(set(y))
Dataset.target_names
```

```python
df["TARGET"]=y
df
```

```python
print(x.shape)
print(y.shape)
```

```python
x_train,x_test,y_train,y_test=train_test_split(x,y,test_size=0.2,random_state=42)
```

```python
x_train_std=StandardScaler().fit_transform(x_train)
x_train_mm=MinMaxScaler().fit_transform(x_train)
```

```python
x_train_std
```

```python
x_train_mm
```

```python
x_test_std=StandardScaler().fit_transform(x_test)
x_test_mm=MinMaxScaler().fit_transform(x_test)
```

```python
x_test_std
```

```python
x_test_mm
```

```python
Flowers=KNeighborsClassifier(n_neighbors=5)
Flowers.fit(x_train,y_train)
```

```python
y_pred=Flowers.predict(x_test)
```

```python
print("\n\nAccuracy:\n\n",accuracy_score(y_test,y_pred))
print("\n\nConfusion Matrix:\n\n",confusion_matrix(y_test,y_pred))
print("\n\nClassification Report:\n\n",classification_report(y_test,y_pred))
```

```python
CFM=confusion_matrix(y_test,y_pred)
Labels=Dataset.target_names
plt.figure(figsize=(6,5))
sns.heatmap(CFM,fmt="d",annot=True,xticklabels=Labels,yticklabels=Labels,cmap="viridis")
plt.title("HeatMap")
plt.xlabel("Predicted Values")
plt.ylabel("Actual Values")
plt.show()
```
