#**DATASPLIT-UP**

```python
x_train,x_test,y_train,y_test=train_test_split(x,y,test_size=0.2,random_state=2)
```

# **DECISION TREE CLASSIFIER**

```python
doctor=DecisionTreeClassifier()
doctor.fit(x_train,y_train)
```

```python
y_pred=doctor.predict(x_test)
```

```python
accuracy=doctor.score(x_test,y_test)
print("ACCURACY:",accuracy)
```

#**CONFUSION MATRIX**

```python
from enum import CONFORM
CFM=confusion_matrix(y_test,y_pred)
CFM
```

```python
CFM=confusion_matrix(y_test,y_pred)
```

#**CLASSIFICATION REPORT**

```python
CFR=classification_report(y_test,y_pred)
CFR
```

#**VISUALIZATION**

```python
Labels=Data.target_names

plt.figure(figsize=(6,5))

sns.heatmap(CFM,annot=True,fmt="d",xticklabels=Labels,yticklabels=Labels,cmap="Oranges")

plt.title("HEATMAP")
plt.xlabel("Predicted Action")
plt.ylabel("Actual Action")
plt.show()
```

```python
plt.figure(figsize=(6,5))
plt.pie(CFM.diagonal(),labels=Labels,autopct="%1.2f%%")
plt.title("CONFUSION MATRIX")
plt.show()
```
